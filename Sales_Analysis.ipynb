### Data Analysis | visulalisation

```python
cnt = sns.countplot(x = 'Age_Group', data = df)
```

```python
a = df['Age_Group'].sort_values(ascending = True)
cnt = sns.countplot(x = 'Age_Group', data = pd.DataFrame(a))
```

```python
cnt = sns.countplot(x = 'Gender', data = df)
```

```python
cnt = sns.countplot(x = 'Gender', data = df)

for bar in cnt.containers:
    cnt.bar_label(bar)
```

```python
group = df.groupby('Gender')['Amount'].sum()
g = pd.DataFrame(group)
g
```

```python
c = sns.barplot(x = 'Gender', y = 'Amount', data = g)
```

```python
gender_counts = df['Gender'].value_counts()
gender_counts
```

```python
plt.pie(gender_counts, labels = gender_counts.index,autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.show()
```

```python
gender_labels = {
    'M': 'Male',
    'F': 'Female'
}
```

```python
labels=[gender_labels[i] for i in gender_counts.index]
labels
```

```python
plt.figure(figsize = (7,7))
plt.pie(gender_counts,
       labels = labels,
        autopct = '%1.1f%%'
       )
plt.show()
```

```python
age_counts = df['Age_Group'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    age_counts,
    labels=age_counts.index,
    autopct='%1.f%%'
)
plt.title('Age Group Distribution')
plt.show()
```

```python
sns.barplot(data = df.groupby('Zone')['Amount'].sum())
plt.title("Total Amount by Zone")
```

```python

zone_amount = df.groupby('Zone')['Amount'].sum().reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(data=zone_amount, x='Zone', y='Amount')

for i in ax.containers:
    ax.bar_label(i, fmt='%.0f')

ax.ticklabel_format(style='plain', axis='y') 

plt.title("Total Amount by Zone")
plt.ylabel("Total Amount")
plt.show()
```
